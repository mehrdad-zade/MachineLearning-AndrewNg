```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('original.csv')
```

```python
x = DataFrame(data, columns = ['production_budget_usd'])
y = DataFrame(data, columns = ['worldwide_gross_usd'])
```

```python
regression = LinearRegression()
regression.fit(x,y)
```

```python
print('theta0 : {0}'.format(regression.intercept_))  #this is theta 0 (constant)
print('theta1 : {0}'.format(regression.coef_)) #this is the slope, theta 1
print('Goodness of fit - r^2 : {0}'.format(regression.score(x,y)*100)) # it should the percentage of variations. with more features it should grow.
```

```python
plt.figure(figsize=(20,6))#increase plot size (width, height)
plt.scatter(x,y);
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.xlim(0, 450000000)#provide proper numbering on the x 
plt.ylim(0, 3000000000)
plt.plot(x, regression.predict(x), color='red', linewidth=4) #depict the linear reg line on the data
plt.show();
```
